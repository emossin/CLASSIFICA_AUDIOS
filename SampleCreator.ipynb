```python
import sounddevice as sd
import wavio as wv

freq = 44100
duration = 5
count=0
while True:
    recording = sd.rec(int(duration * freq), samplerate=freq, channels=2)
    sd.wait()
    #write("recording0.wav", freq, recording)
    wv.write("./base/gravados1/silencio" + str(count) + ".wav", recording, freq, sampwidth=2)
    print("\ndone " + "recording" + str(count) + ".wav")
    count = count+1
```
